```python
from sklearn.tree import DecisionTreeClassifier
```

```python
fruits =  [[150,8],[170,9],[130,7],[200,3],[210,2],[190,4],[80,6],[90,7],[85,8]]
labels =  ['Apple','Apple','Apple','Orange','Orange','Orange','Grape','Grape','Grape']

model = DecisionTreeClassifier()
model.fit(fruits, labels)

weight = float(input("Enter fruit weight (grams): "))
sweetness = float(input("Enter sweetness (1-10): "))

prediction = model.predict([[weight, sweetness]])[0]
print(f"Predicted Fruit:  {prediction}")

```
