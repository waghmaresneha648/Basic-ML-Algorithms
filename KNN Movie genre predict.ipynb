```python
from sklearn.neighbors import KNeighborsClassifier

# Features: [action_scenes, romance_scenes]
movies = [[9,1],[8,2],[7,1],[1,9],[2,8],[1,7],[5,5],[4,6],[6,4]]
genres = ['Action','Action','Action','Romance','Romance','Romance','Comedy','Comedy','Comedy']

model = KNeighborsClassifier(n_neighbors=3)
model.fit(movies, genres)

action = float(input("Enter action scenes (1-10): "))
romance = float(input("Enter romance scenes (1-10): "))

prediction = model.predict([[action, romance]])[0]
print(f"Predicted Genre: ðŸŽ¬ {prediction}")
```
